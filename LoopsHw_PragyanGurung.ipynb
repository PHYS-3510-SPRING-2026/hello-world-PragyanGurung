```python
#Question 2
#The functions:
#a(x)=4x^2+3
#b(x)=7x^3-4x
#c(x)=ln(x^2+5)-6x^2

import numpy as np
x=3
h_values=[0.01,0.001,0.0001]
for h in h_values:
  print("for h =",h)

  #function a
  a_xh=4*(x+h)**2+3
  a_x=4*x**2+3
  a_prime=(a_xh-a_x)/h
  print("Value of a at 3 is",a_prime)

  #function b
  b_xh=7*(x+h)**3-4*(x+h)
  b_x=7*x**3-4*x
  b_prime=(b_xh-b_x)/h
  print("Value of b at 3 is",b_prime)

  #function c
  c_xh=np.log((x+h)**2+5)-6*(x+h)**2
  c_x=np.log(x**2+5)-6*x**2
  c_prime=(c_xh-c_x)/h
  print("Value of c at 3 is",c_prime)
  print("                         ")


```
