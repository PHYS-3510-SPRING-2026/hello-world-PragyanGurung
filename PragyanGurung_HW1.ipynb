```python
#Problem 1 Gauss' Sum
#I choose the sequence from 25 to 30
#Using Brute force
a=25
a=a+26
a=a+27
a=a+28
a=a+29
a=a+30
print("The sum of the sequence using brute force is",a)

#Using Gauss's Sum
#Assigning variables
n=6
F=25
L=30
#Using the formula
b=n*(F+L)/2
print("The sum of the sequence using Gauss' formula is", int(b))
c=(a==b)
if c==True:
  print("The two sums are equal.")
```
